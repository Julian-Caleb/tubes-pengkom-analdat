```python
import pandas as pd
import matplotlib.pyplot as plt

phones = pd.read_csv("phones_data.csv")
phones = phones.dropna(subset=["os", "lowest_price", "highest_price", "battery_size"])
```

Perbandingan Kategori

```python
phones["brand_name"].value_counts().plot(kind="bar",title="Jumlah HP Untuk Setiap Nama Brand",xlabel="Nama Brand",ylabel="Fekuensi HP");
# x = nama brand, y = jumlah hp setiap brand
```

```python
phones["os"].value_counts().plot(kind="bar", title="OS Yang Digunakkan",xlabel="Operating System",ylabel="Frekuensi HP");
# x = nama operating system, y = jumlah hp yang menggunakan tiap operating system
```

Penampilan Perubahan Terhadap Waktu

```python
phones.plot(kind="area",x="release_date",y=["lowest_price", "best_price", "highest_price"], title="Perubahan Harga Terhadap Waktu", ylabel="Harga");
# x = waktu/tanggal, y = harga
```

```python
phones.plot(kind="area",x="release_date",y="sellers_amount", title="Perubahan Jumlah Penjual Terhadap Waktu", ylabel="Frekuensi Penjual");
# x = waktu/tanggal, y = jumlah penjual
```

```python
phones.plot(kind="line",x="release_date",y="screen_size", title="Perubahan Ukuran Layar Terhadap Waktu",ylabel="Ukuran Layar");
# x = waktu/tanggal, y = ukuran layar
```

```python
phones.plot(kind="line",x="release_date",y="memory_size", title="Perubahan Ukuran Memori Terhadap Waktu",ylabel="Ukuran Memori");
# x = waktu/tanggal, y = ukuran memori
```

```python
phones.plot(kind="area",x="release_date",y="battery_size", title="Perubahan Ukuran Baterai Terhadap Waktu",ylabel="Ukuran Baterai");
# x = waktu/tanggal, y = ukuran memori
```

Penampilan Hierarki dan Hubungan Keseluruhan Bagian

```python
phones.groupby(["os","brand_name"]).size().unstack().plot(kind="barh",stacked=True, title="Banyaknya Masing-Masing Brand Untuk Setiap Operating System");
plt.legend(ncol = 3);
```

```python
data = phones.loc[phones["os"] == "Android"]
data["brand_name"].value_counts().plot(kind="pie", title="Brand Name Yang Menggunakan Operating System Android");
```

```python
data = phones.loc[phones["os"] == "iOS"]
data["brand_name"].value_counts().plot(kind="pie", title="Brand Name Yang Menggunakan Operating System iOS");
```

Plotting Relationship

```python
phones.plot(kind="scatter", x="screen_size", y="best_price", title="Plot Ukuran Layar dan Harga Terbaik");
phones["screen_size"].corr(phones["best_price"])
# x = ukuran layar, y = harga terbaik
```
